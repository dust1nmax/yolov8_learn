```python
video = cv2.VideoCapture("League of Legends (TM) Client 2024-07-20 11-51-12.mp4")
num = 0
save_step = 30
while True:
    ret, frame = video.read()
    if not ret:
        break
    num += 1
    if num % save_step == 0:
        cv2.imwrite("./demo_images/" + str(num) + ".jpg", frame) #按帧从视频里抽出来 放在demo_images里
```
